```python
# Ask the user for a temperature like 37C or 98.6F and print the converted value in the other unit.”

temp = input("Enter the temperature: ").strip().upper()


if temp.endswith("C"):
    c = float(temp[:-1])
    f = c * 9/5 + 32
    print(f"{c:.1f}°C = {f:.1f}°F")

elif temp.endswith("F"):
    f = float(temp[:-1])
    c = (f - 32) * 5/9
    print(f"{f:.1f}°F = {c:.1f}°C")

else:
    print("Please enter a valid format (e.g., 37C or 98.6F)")
```

```python
#Simple Calculator (+, −, ×, ÷)
#“Ask for two numbers and an operator (+, -, *, /). Print the result. Handle bad input and divide-by-zero.”

num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "*":
    print(num1 * num2)
elif op == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero")
    else:
        print(num1 / num2)
else:
    print("Invalid operator")
```

```python
#Email Slicer
#“Ask the user for an email address and print the username and the domain. Also show the top-level domain if present (like com, edu).”

email = input("Enter your email: ").strip().lower()
if "@" in email:
    parts = email.split("@")
    username = parts[0]
    domain = parts[1]

    print("Username:", username)
    print("Domain:", domain)
    if "." in domain:
        tld = domain.split(".")[-1]
        print("Top-level domain:", tld)
else:
    print("Invalid email format")
```

```python
#Expense Splitter
#“Given N friends and how much each one paid, show the total, the fair per-person amount, and how much each person owes or should receive

n = int(input("How many friends? "))
names = []
amounts = []

for i in range(n):
    name = input(f"Name of friend #{i+1}: ")
    paid = float(input(f"Amount paid by {name}: "))
    names.append(name)
    amounts.append(paid)

total = sum(amounts)
per_head = total / n

print("\nTotal:", total)
print("Each should pay:", per_head)

for i in range(n):
    diff = amounts[i] - per_head
    if diff == 0:
        print(names[i], "is settled")
    elif diff > 0:
        print(names[i], "should receive", diff)
    else:
        print(names[i], "owes", -diff)
```

```python
#“Generate a random password of a user-chosen length (8–20) that contains at least one uppercase, one lowercase, one digit, and one symbol.”

import random
import string
length = int(input("Enter password length (8-20): "))

if length < 8 or length > 20:
    print("Length must be between 8 and 20")
else:
    uppercase = string.ascii_uppercase
    lowercase = string.ascii_lowercase
    digits = string.digits
    symbols = "!@#$%^&*()-_=+[]{};:,.?/"

    password_chars = [
        random.choice(uppercase),
        random.choice(lowercase),
        random.choice(digits),
        random.choice(symbols)
    ]

    all_chars = uppercase + lowercase + digits + symbols
    remaining = length - len(password_chars)
    for i in range(remaining):
        password_chars.append(random.choice(all_chars))
    random.shuffle(password_chars)
    password = "".join(password_chars)

    print("Generated password:", password)
```
